## Convolution

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt # library to make plot

# Charger l'image
image = cv2.imread("esgi.jpg")


# Définir le noyau (filtre ou kernel) de convolution
kernel = np.array([[0, 1, 10],
                   [0, 1, 10],
                   [0, 1, 10]])

# Appliquer la convolution à l'image
convolution = cv2.filter2D(image, -1, kernel)

# Afficher les images côte à côte
fig, axes = plt.subplots(1, 2, figsize=(15, 15))  # Ajuster la taille ici (largeur, hauteur)

# Afficher l'image d'origine
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Image originale')

# Afficher l'image après convolution
axes[1].imshow(cv2.cvtColor(convolution, cv2.COLOR_BGR2RGB))
axes[1].set_title('Image après convolution')

# Masquer les graduations et les axes
for ax in axes:
    ax.axis('off')

# Ajuster la disposition pour éviter les superpositions
plt.tight_layout()

# Afficher le graphique
plt.show()
```

## Maxpooling

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image
image = cv2.imread("esgi.jpg")


# Définir la taille du noyau de max pooling et le pas (stride)
compression=10

# Appliquer le max pooling à l'image
maxpool = cv2.resize(image, (image.shape[1]//compression, image.shape[0]//compression))

# Ajuster la taille d'affichage
fig, axes = plt.subplots(1, 2, figsize=(14, 8))  # Ajuster la taille ici (largeur, hauteur)

# Afficher l'image d'origine
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Image originale')


# Afficher l'image après max pooling
axes[1].imshow(cv2.cvtColor(maxpool, cv2.COLOR_BGR2RGB))
axes[1].set_title('Image après max pooling')

# Ajuster la disposition pour éviter les superpositions
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Chemin vers l'image
image_path = "esgi.jpg"

# Charger l'image
image = cv2.imread(image_path)

# Définir les matrices de convolution célèbres
laplacian_kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpen_kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
horizontal_edges_kernel = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
vertical_edges_kernel = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
blur_kernel = np.array([[0.0625, 0.125, 0.0625], [0.125, 0.25, 0.125], [0.0625, 0.125, 0.0625]])
emboss_kernel = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]])

# Liste des filtres de convolution avec leurs noms correspondants
filters = [
    (laplacian_kernel, "Filtre de Laplace"),
    (sharpen_kernel, "Filtre de netteté"),
    (horizontal_edges_kernel, "Filtre de détection de contours horizontaux"),
    (vertical_edges_kernel, "Filtre de détection de contours verticaux"),
    (blur_kernel, "Filtre de flou"),
    (emboss_kernel, "Filtre d'embossage")]

# Afficher les images résultantes des convolutions en couleur
plt.figure(figsize=(12, 10))

for i, (kernel, filter_name) in enumerate(filters):
    # Appliquer la convolution
    convolved_image = cv2.filter2D(image, -1, kernel)

    # Afficher l'image résultante avec le nom du filtre
    plt.subplot(4, 2, i+1)
    plt.imshow(cv2.cvtColor(convolved_image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title(filter_name)

plt.tight_layout()
plt.show()
```

# Préparation des modèles - Step By Step

Nos données étant désormais prêtes, nous allons pouvoir les utiliser au sein d'un modèle de type "Convolutional neural network" (CNN), ce dernier va reproduire un réseau de neurones artificiels qui est spécialement adapté au traitement des images. Ce réseau est inspiré des processus biologiques et les scientifiques ont remarqué que les méthodes qui fonctionnent avec les véritables réseaux de neuronnes s'appliquent également aux réseaux artificiels.

Notre objectif est d'obtenir un modèle avec une précision trés élevée, nous allons démarrer avec un modèle disposant d'un nombre faible de couches, puis nous ferons varier les différents paramètres pour augmenter nos résultats.

## Modèle n°0

```python
#day 1
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten

# Initial the model as a Sequential model
model_0 = Sequential()

# Add a flatten layer
model_0.add(Flatten())

# Output layer (densely connected layer)
model_0.add(Dense(2, activation='softmax')) # 2 outputs with activation function relu

model_0.compile(optimizer='adam', 
                loss='categorical_crossentropy', 
                metrics=['accuracy'])

model_0_fit_output = model_0.fit(training_images, 
                                 epochs=10, 
                                 batch_size=64,
                                 validation_data=testing_images)
```

## Modèle n°1

Nous allons effectuer du preprocessing sur nos données afin de les optimiser pour un traitement par le modèle.
Pour cela nous allons convertir le format actuel des données représentant les couleurs des images (les trois canaux) étendus sur 255 vers un format binaire, ce type de données étant plus adapté pour le modèle.

Notre modèle va suivre la structure générale : 
 convolutional layers, pooling layers, dropout layers, fully connected hidden layers.

la représentation graphique de notre modèle nous permet de visualiser les résultats des opérations sur les matrices.

Une fois notre modèle défini, nous pouvons le compiler et commencer à l'entrainer. 

Nous définissons le nombre d'apprentissage que nous souhaitons. Pour notre premier modèle, nous allons le garder court et vérifier nos premiers résultats.

Nous compilons ensuite notre modèle avec l'Optimizer Adam et nous précisons que nous utilisons la "sparse_categorical_crossentropy" car nous avons des données multiclasse dispersées.

Enfin, nous allons entrainer notre modèle et le valider.


```python
#day 1
from keras.models import Sequential  #couche1 couche2 .....derniere couche
from keras.layers import Dense  #toutes les couches sont connectées

#day2
#info:https://developersbreach.com/convolution-neural-network-deep-learning/

from keras.layers import Conv2D, MaxPooling2D, Flatten  #filter image, reduire image, applatir image

# Initial the model as a Sequential model
model_1 = Sequential()

# Add a convolution Conv2D layer
#livre: http://exo7.emath.fr/cours/livre-deepmath.pdf

#TRAITEMENT DES IMAGES
model_1.add(Conv2D(filters=24, # dimentionality of the output space (on passe d'1 à 24 images)
                 kernel_size=(3,3), # height and width of the 2D convolution window
                 activation='relu', # the activation function (here it's relu)
                 input_shape=(64, 64, 3))) # input dimensions, 64x64 with RGB       #24*64*64*3

# Add a max pooling layer
model_1.add(MaxPooling2D(pool_size=(2,2))) # Apply Max Pooling with a 2x2 matrix (get the max value of the 4 matrix value)
#effet du maxpoling:https://blog.paperspace.com/pooling-in-convolutional-neural-networks/
#FIN TRAITEMENT

# Add a flatten layer
model_1.add(Flatten())

#on repete le cours du day2
# Add a densely connected layer
model_1.add(Dense(100, activation='relu')) # 20 outputs with activation function relu

# Output layer (densely connected layer)
model_1.add(Dense(2, activation='softmax')) # 2 outputs with activation function relu

#ici on optimise les erreurs: par exemple le modele doit tout faire pour trouver les labels de toutes les images
model_1.compile(optimizer='adam', 
                loss='categorical_crossentropy', 
                metrics=['accuracy'])

model_1_fit_output = model_1.fit(training_images, 
                                 epochs=10, 
                                 batch_size=64,
                                 validation_data=testing_images)
```

Nous pouvons voir que nos premiers résultats sont loin d'être exceptionnels, nous relevons 97% pour l entrainement et 62% pour la validation ou test (conv+pool+100+2)

Toutefois, nous pouvons voir que nous avons encore une large marge d'amélioration avec ce modèle en l'état en le faisant apprendre plus longtemps, notre valeur de loss étant encore trés élevé.

## Modèle n°2

Nous allons augmenter nos pas et ajouter des couches en entrée de notre modèle.

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten, BatchNormalization

# Initial the model as a Sequential model
model_2 = Sequential()

# Convolutional Conv2D + Batch Normalization + Dense layers
model_2.add(Conv2D(filters=30, # dimentionality of the output space 
                 kernel_size=(3,3), # height and width of the 2D convolution window
                 activation='relu', # the activation function (here it's relu)
                 padding='same', # same padding as input
                 input_shape=(64, 64, 3))) # input dimensions, 64x64 with RGB
model_2.add(MaxPooling2D(pool_size=(2,2))) # Apply Max Pooling with a 2x2 matrix (get the max value of the 4 matrix value)

# Convolutional Conv2D + Batch Normalization + Dense layers
model_2.add(Conv2D(filters=6, # dimentionality of the output space 
                 kernel_size=(3,3), # height and width of the 2D convolution window
                 activation='relu', # the activation function (here it's relu)
                 padding='same', # same padding as input
                 input_shape=(32, 32, 3))) # input dimensions, 64x64 with RGB
model_2.add(MaxPooling2D(pool_size=(2,2))) # Apply Max Pooling with a 2x2 matrix (get the max value of the 4 matrix value)




# Flatten layer
model_2.add(Flatten())

model_2.add(Dense(64, activation='relu')) 

#Normalisation data
model_2.add(BatchNormalization())   #ici on normalise par lot les donnees pour qu'elles soient aux memes echelles


# Dropout layer
model_2.add(Dropout(0.05)) # Drop 5% of the neurons/nodes: par exemple si vous avez un modele qui s'entraine à 95% par contre en test il a une performance de 60%. conclusion: ce modele est trop complexe pour notre probeleme et il faut donc enlever certains neurones

# Densely connected layer
model_2.add(Dense(30, activation='relu')) # 30 outputs with activation function relu

# Output layer (densely connected layer)
model_2.add(Dense(2, activation='softmax')) # 10 outputs with activation function relu
model_2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_2_fit_output = model_2.fit(
    training_images, 
    epochs=15, 
    batch_size=64,
    validation_data=testing_images
)
```

```python
visualkeras.layered_view(model_2,legend=True)
```

Nous pouvons de nouveau constater une amélioration de 4% sur nos données d'entrainement et d'environ 10% sur nos données de test. Nous allons effectuer quelques ajouts de couches supplémentaires avant de commencer à nous entrainer sur des durées plus longues.

Nos résultats sont bien meilleurs, nous décidons de retenir ce modèle.

## 8. Bonus: use a pre-trained model like ResNet to improve your results.

ResNet is a pre-trained model which won the ILSVRC in 2015. It is made of 152 layers.

We need to import the ResNet model to use it:

```python
from tensorflow.keras.applications import ResNet50
```

We create a new model from the ResNet50 model and predict the content of an image of a guinea pig found on Google Images:

```python
import numpy as np
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions

# Load the ResNet50 pre-trained model
model_resnet = ResNet50(weights='imagenet')

# Guinea pig image filename
image_file = 'cochon-inde.jpg'

# Load and resize the guinea pig image
image=preprocessing.image.load_img(image_file, target_size=(224, 224)) # ResNet50 use 224x244 resolution

# Transform the image to an array (preprocessing)
image_array = preprocessing.image.img_to_array(image)

# Expand the image dimensions (preprocessing)
image_tensor= tensorflow.expand_dims(image_array, axis=0)

# Use the ResNet50 preprocess_input function, conseillez pour les mettre aux formats standard de resnet
image_resnet = preprocess_input(image_tensor)

# Make the predictions using the ResNet50 model
predictions_resnet = model_resnet.predict(image_resnet)

# Print the 5 most probable object on the image using the ResNet50 decode_predictions function
decode_predictions(predictions_resnet, top=5)[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
corpus = [
     'This is the first document.',
     'This document is the second document.',
     'And this is the third one.',
     'Is this the first document?']
corpus
```

```python
vectorizer = CountVectorizer(ngram_range=(2, 2))
```

```python
X = vectorizer.fit_transform(corpus)  #on transforme notre corpus en numerique
X
```

```python
print(X.toarray())
```

```python
vectorizer.get_feature_names_out()
```

```python
import pandas
data=pandas.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
data
```
