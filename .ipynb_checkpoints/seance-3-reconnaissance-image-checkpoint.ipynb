# Pr√©paration des mod√®les - Step By Step

Nos donn√©es √©tant d√©sormais pr√™tes, nous allons pouvoir les utiliser au sein d'un mod√®le de type "Convolutional neural network" (CNN), ce dernier va reproduire un r√©seau de neurones artificiels qui est sp√©cialement adapt√© au traitement des images. Ce r√©seau est inspir√© des processus biologiques et les scientifiques ont remarqu√© que les m√©thodes qui fonctionnent avec les v√©ritables r√©seaux de neuronnes s'appliquent √©galement aux r√©seaux artificiels.

## Mod√®le n¬∞1

Nous allons effectuer du preprocessing sur nos donn√©es afin de les optimiser pour un traitement par le mod√®le.
Pour cela nous allons convertir le format actuel des donn√©es repr√©sentant les couleurs des images (les trois canaux) √©tendus sur 255 vers un format binaire, ce type de donn√©es √©tant plus adapt√© pour le mod√®le.

Notre mod√®le va suivre la structure g√©n√©rale : 
 convolutional layers, pooling layers, dropout layers, fully connected hidden layers.

la repr√©sentation graphique de notre mod√®le nous permet de visualiser les r√©sultats des op√©rations sur les matrices.

Une fois notre mod√®le d√©fini, nous pouvons le compiler et commencer √† l'entrainer. 

Nous d√©finissons le nombre d'apprentissage que nous souhaitons. Pour notre premier mod√®le, nous allons le garder court et v√©rifier nos premiers r√©sultats.

Nous compilons ensuite notre mod√®le avec l'Optimizer Adam et nous pr√©cisons que nous utilisons la "sparse_categorical_crossentropy" car nous avons des donn√©es multiclasse dispers√©es.

Enfin, nous allons entrainer notre mod√®le et le valider.


```python
#day 1
from keras.models import Sequential
from keras.layers import Dense

#day2
#info:https://developersbreach.com/convolution-neural-network-deep-learning/

from keras.layers import Conv2D, MaxPooling2D, Flatten

# Initial the model as a Sequential model
model_1 = Sequential()

# Add a convolution Conv2D layer
#livre: http://exo7.emath.fr/cours/livre-deepmath.pdf

#TRAITEMENT DES IMAGES
model_1.add(Conv2D(filters=24, # dimentionality of the output space (on passe d'1 √† 24 images)
                 kernel_size=(3,3), # height and width of the 2D convolution window
                 activation='relu', # the activation function (here it's relu)
                 input_shape=(64, 64, 3))) # input dimensions, 64x64 with RGB       #24*64*64*3

# Add a max pooling layer
model_1.add(MaxPooling2D(pool_size=(2,2))) # Apply Max Pooling with a 2x2 matrix (get the max value of the 4 matrix value)
#effet du maxpoling:https://blog.paperspace.com/pooling-in-convolutional-neural-networks/
#FIN TRAITEMENT

# Add a flatten layer
model_1.add(Flatten())

#on repete le cours du day2
# Add a densely connected layer
model_1.add(Dense(110, activation='relu')) # 60 outputs with activation function relu
model_1.add(Dense(50, activation='relu')) # 60 outputs with activation function relu
model_1.add(Dense(5, activation='relu')) # 60 outputs with activation function relu

# Output layer (densely connected layer)
model_1.add(Dense(2, activation='softmax')) # 2 outputs with activation function relu

model_1.compile(optimizer='adam', 
                loss='categorical_crossentropy', 
                metrics=['accuracy'])

model_1_fit_output = model_1.fit(training_images, 
                                 epochs=10, 
                                 batch_size=64,
                                 validation_data=testing_images)
```

Nous pouvons voir que nos premiers r√©sultats sont loin d'√™tre exceptionnels, nous relevons 88% pour l entrainement et 68% pour la validation ou test.

Toutefois, nous pouvons voir que nous avons encore une large marge d'am√©lioration avec ce mod√®le en l'√©tat en le faisant apprendre plus longtemps, notre valeur de loss √©tant encore tr√©s √©lev√©.

## Mod√®le n¬∞2

Nous allons augmenter nos pas et ajouter des couches en entr√©e de notre mod√®le.

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, Flatten, BatchNormalization

# Initial the model as a Sequential model
model_2 = Sequential()

# Convolutional Conv2D + Batch Normalization + Dense layers
model_2.add(Conv2D(filters=32, # dimentionality of the output space 
                 kernel_size=(3,3), # height and width of the 2D convolution window
                 activation='relu', # the activation function (here it's relu)
                 padding='same', # same padding as input
                 input_shape=(64, 64, 3))) # input dimensions, 64x64 with RGB
model_2.add(MaxPooling2D(pool_size=(2,2))) # Apply Max Pooling with a 2x2 matrix (get the max value of the 4 matrix value)

# Flatten layer
model_2.add(Flatten())

model_2.add(Dense(64, activation='relu')) 

#Normalisation data
model_2.add(BatchNormalization())


# Dropout layer
model_2.add(Dropout(0.05)) # Drop 5% of the neurons/nodes

# Densely connected layer
model_2.add(Dense(30, activation='relu')) # 30 outputs with activation function relu

# Output layer (densely connected layer)
model_2.add(Dense(2, activation='softmax')) # 10 outputs with activation function relu
model_2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_2_fit_output = model_2.fit(
    training_images, 
    epochs=15, 
    batch_size=64,
    validation_data=testing_images
)
```

Nous pouvons de nouveau constater une am√©lioration de 4% sur nos donn√©es d'entrainement et d'environ 10% sur nos donn√©es de test. Nous allons effectuer quelques ajouts de couches suppl√©mentaires avant de commencer √† nous entrainer sur des dur√©es plus longues.

Nos r√©sultats sont bien meilleurs, nous d√©cidons de retenir ce mod√®le.

## 7. Use your model with your own photos or some photos find on internet.

```python
from tensorflow.keras import preprocessing
import tensorflow
import keras
# Animal classes/labels to interpret results
animal_classes = ['cat','guinea_pig']

# Guinea pig image filename
image_file = 'cochon-inde.jpg'

# Load and resize the guinea pig image with the Keras load_img function
image_to_predict_with_model_2 = preprocessing.image.load_img(image_file, target_size=(64,64))

# Convert the image to an array
img_array = preprocessing.image.img_to_array(image_to_predict_with_model_2)

# Add an axis to allow the model to use the image to predict (else it's not well formatted)
img_array = tensorflow.expand_dims(img_array, 0)  # Create batch axis

# Run the prediction using the previously defined model
predictions_model_2 = model_2.predict(img_array)

for i in range(nbclass): # for each 10 classes (animals)
    print(animal_classes[i] + ": " + str(predictions_model_2[0][i])) # Print the probability for each results
```

We can see the results, and the model has well determined what is the animal (a guinea pig) on the image found on Google Images.

## 8. Bonus: use a pre-trained model like ResNet to improve your results.

ResNet is a pre-trained model which won the ILSVRC in 2015. It is made of 152 layers.

We need to import the ResNet model to use it:

```python
from tensorflow.keras.applications import ResNet50
```

We create a new model from the ResNet50 model and predict the content of an image of a guinea pig found on Google Images:

```python
import numpy as np
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions

# Load the ResNet50 pre-trained model
model_resnet = ResNet50(weights='imagenet')

# Guinea pig image filename
image_file = 'cochon-inde.jpg'

# Load and resize the guinea pig image
image_to_predict_with_resnet=preprocessing.image.load_img(image_file, target_size=(224, 224)) # ResNet50 use 224x244 resolution

# Transform the image to an array (preprocessing)
image_to_predict_with_resnet_preprocessing = preprocessing.image.img_to_array(image_to_predict_with_resnet)

# Expand the image dimensions (preprocessing)
image_to_predict_with_resnet_preprocessing = np.expand_dims(image_to_predict_with_resnet_preprocessing, axis=0)

# Use the ResNet50 preprocess_input function
image_to_predict_with_resnet_preprocessing = preprocess_input(image_to_predict_with_resnet_preprocessing)

# Make the predictions using the ResNet50 model
predictions_resnet = model_resnet.predict(image_to_predict_with_resnet_preprocessing)

# Print the 5 most probable object on the image using the ResNet50 decode_predictions function
print('Predicted:', decode_predictions(predictions_resnet, top=5)[0])
```

In the result we can read that the prediction is:
* 99,598986% guinea_pig (good prediction)
* 0.15480462% hamster
* 0.078452955% bath_towel
* 0.05238239% broccoli
* 0.018165231% hare

So, predictions with the model ResNet50 are very efficient, and this illustrate how deep learning can be powerful to identify elements on images (for example).
